# Hash tables

**Most of the problems in code interviews are solved with hash tables**.

* Hash tables are a type of data structure in which the address or the index value of the data element is generated from a hash function. **That makes accessing the data faster as the index value behaves as a key for the data value**.


* *In other words Hash table sores key-value pairs but the key is generated through a hashing function*

In python, the *dictionary data types* represent the implementation of hash tables. The Keys in the dictionary satisfy the following requirements:
* The **key** of dictionary are hashable i.e they are **generated by hashing function which generates unique value suplied to the hash function**
* The order of adata elements in a dictionary in not fixed


```python
# Declaring a dictionary
dict = {'Name': 'Zara', 'Age': 7, 'Class ': 'First'}

# Accessing the dictionary with its key
print ("Dict['Name']: ", dict['Name'])
print ("Dict['Age']: ", dict['Age'])
```

```python
#Updating Dictionary
dict['Age'] = 8
#Adding a new entry
dict['School'] = "DPS School"

print ("Dict['School']: ", dict['School'])
```

```python
print(dict)
```

**Removing elements**

```python
dict = {'Name': 'Zara', 'Age': 7, 'Class ': 'First'}
# Removing entry with key 'Name'
del dict['Name']
print (dict)
```

```python
# Remove all entries in dict
dict.clear()
print(dict)
```

```python
# Delete entire dictionary
del dict
print(dict)
```

**Check if Key Exists**

```python
thisdict = {
        'Name': 'Hugo', 
        'Age': 21, 
        'Class ': 'FCE'
}
```

```python
if 'Age' in thisdict:
    print("Yes, model is one of the keys in the dictionary")
```

**Making a copy**

```python
mydict = dict(thisdict)
print(mydict)
```

```python
thisdict = dict(Name = 'Hugo', Age = 21, Class = 'FCE')

# Note that keywords are not string literals
# note the use of equals rather than colon(:) for the assigment
```
